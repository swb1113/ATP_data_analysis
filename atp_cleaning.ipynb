#### 5b-1. Converting columns w/ 2 data types --> Rank Columns

```python
# renaming columns to split strings
df.rename(columns = {'Best Elo Rank':'Best_Elo_Rank', 'Best Rank':'Best_Rank', 'Current Elo Rank':'Current_Elo_Rank', 'Current Rank': 'Current_Rank', 'GOAT Rank': 'GOAT_Rank'}, inplace = True)

# splitting columns
df[['Best_Elo_Rank', 'Best_Elo_Rank_Date']] = df.Best_Elo_Rank.str.split("(", expand = True)
df[['Best_Rank', 'Best_Rank_Date']] = df.Best_Rank.str.split("(", expand = True)
df[['Current_Elo_Rank', 'Current_Elo_Rank_Points']] = df.Current_Elo_Rank.str.split("(", expand = True)
df[['Current_Rank', 'Current_Rank_Points']] = df.Current_Rank.str.split("(", expand = True)
df[['GOAT_Rank', 'GOAT_Rank_Points']] = df.GOAT_Rank.str.split("(", expand = True)
```

```python
# replacing NaN values with '0' to convert rank column into int64

df[['GOAT_Rank', 'GOAT_Rank_Points']] = df[['GOAT_Rank', 'GOAT_Rank_Points']].fillna('0')
```

```python
# converting Rank_Points columns into int64
for column in df:
    if 'Rank_Points' in column:
        df[column] = df[column].str.rstrip(")").astype('int64')
```

```python
# converting Date columns into datetime 
df['Best_Rank_Date'] = df['Best_Rank_Date'].str.rstrip(")")
df['Best_Rank_Date'] = pd.to_datetime(df['Best_Rank_Date'], format= "%d-%m-%Y")

df['Best_Elo_Rank_Date'] = df['Best_Elo_Rank_Date'].str.rstrip(")")
df['Best_Elo_Rank_Date'] = pd.to_datetime(df['Best_Elo_Rank_Date'], format= "%d-%m-%Y")
```

```python
# converting rest of Rank columns into int64
df[['Best_Elo_Rank','Best_Rank','Current_Elo_Rank','Current_Rank','GOAT_Rank']] = df[['Best_Elo_Rank','Best_Rank','Current_Elo_Rank','Current_Rank','GOAT_Rank']].astype('int64')
```

```python
df.info()
```

#### 5b-2. Converting columns w/ 2 data types --> Age, Overall, Favorite Surface:

Age: Age (int64), Date_Of_Birth (datetime)  
Overall: Total_Win_Percentage (float), Total_Matches_Won (int64), Total_Matches_Lost (int64)  
Favorite Surface: Favorite_Surface (str), Favorite_Surface_Percentage (float)

```python
# splitting Age and Date_of_Birth
df[['Age', 'Date_Of_Birth']] = df.Age.str.split("(", expand = True)

# converting Age to Int64
df['Age'] = df['Age'].astype('Int64')

# converting Date_Of_Birth to Datetime
df['Date_Of_Birth'] = df['Date_Of_Birth'].str.rstrip(")")
df['Date_Of_Birth'] = pd.to_datetime(df['Date_Of_Birth'], format= "%d-%m-%Y")
```

```python
df.info()
```

```python
# splitting Overall to Total_Win_Percentage, Total_Matches_Won, Total_Matches_Lost 
df[['Total_Win_Percentage', 'Match_Record']] = df.Overall.str.split("(", expand = True)
df[['Total_Matches_Won', 'Total_Matches_Lost']] = df.Match_Record.str.split("-", expand = True)

# converting to appropriate datatypes
df['Total_Matches_Won'] = df['Total_Matches_Won'].astype("int64")
df['Total_Matches_Lost'] = df['Total_Matches_Lost'].str.rstrip(")").astype("int64")
df['Total_Win_Percentage'] = df['Total_Win_Percentage'].str.replace("%","").astype("float")/100
```

```python
# dropping Overall and Match_Record
df = df.drop(['Overall', 'Match_Record'], axis=1)
```

```python
df.info()
```

```python
# renaming Favorite Surface to Favorite_Surface to be able to split
df.rename(columns = {'Favorite Surface':'Favorite_Surface'}, inplace = True)
```

```python
# splitting Favorite_Surface to Favorite_Surface and Favorite_Surface_%
df[["Favorite_Surface", "Favorite_Surface_%"]] = df['Favorite_Surface'].str.partition(' ')[[0, 2]]
#df.head(10)
```

```python
# removing all non-digits in Favorite_Surface_%
df['Favorite_Surface_%'] = df['Favorite_Surface_%'].str.replace('\D', '', regex=True)
```

```python
# renaming cells w/ Fast, Slow, Soft to Fast (H, G), Slow (H, Cl), Soft (Cl, G)

df['Favorite_Surface'] = df['Favorite_Surface'].replace('Fast', 'Fast (H, G)')
df['Favorite_Surface'] = df['Favorite_Surface'].replace('Slow', 'Slow (H, Cl)')
df['Favorite_Surface'] = df['Favorite_Surface'].replace('Soft', 'Soft (Cl, G)')
```

```python
# converting Favorite_Surface_% to float

df['Favorite_Surface_%'] = df['Favorite_Surface_%'].astype(float)/100
```

#### 5b-3. Converting columns w/ 2 data types --> Court Surfaces (Hard, Clay, Carpet, Grass)

```python
df[['Hard', 'Clay', 'Carpet', 'Grass']] = df[['Hard', 'Clay', 'Carpet', 'Grass']].fillna('0% (0-0)')

df.head()
```

```python
# splitting Hard to Hard_Win_Percentage, Hard_Matches_Won, Hard_Matches_Lost 
df[['Hard_Win_Percentage', 'Hard_Record']] = df.Hard.str.split("(", expand = True)
df[['Hard_Matches_Won', 'Hard_Matches_Lost']] = df.Hard_Record.str.split("-", expand = True)

# splitting Clay to Clay_Win_Percentage, Clay_Matches_Won, Clay_Matches_Lost 
df[['Clay_Win_Percentage', 'Clay_Record']] = df.Clay.str.split("(", expand = True)
df[['Clay_Matches_Won', 'Clay_Matches_Lost']] = df.Clay_Record.str.split("-", expand = True)

# splitting Carpet to Carpet_Win_Percentage, Carpet_Matches_Won, Carpet_Matches_Lost 
df[['Carpet_Win_Percentage', 'Carpet_Record']] = df.Carpet.str.split("(", expand = True)
df[['Carpet_Matches_Won', 'Carpet_Matches_Lost']] = df.Carpet_Record.str.split("-", expand = True)

# splitting Grass to Grass_Win_Percentage, Grass_Matches_Won, Grass_Matches_Lost 
df[['Grass_Win_Percentage', 'Grass_Record']] = df.Grass.str.split("(", expand = True)
df[['Grass_Matches_Won', 'Grass_Matches_Lost']] = df.Grass_Record.str.split("-", expand = True)
```

```python
df.iloc[[10,97]]
```

```python
# converting to appropriate datatypes
df['Hard_Matches_Won'] = df['Hard_Matches_Won'].astype("int64")
df['Hard_Matches_Lost'] = df['Hard_Matches_Lost'].str.rstrip(")").astype("int64")
df['Hard_Win_Percentage'] = df['Hard_Win_Percentage'].str.replace("%","").astype("float")/100

df['Clay_Matches_Won'] = df['Clay_Matches_Won'].astype("int64")
df['Clay_Matches_Lost'] = df['Clay_Matches_Lost'].str.rstrip(")").astype("int64")
df['Clay_Win_Percentage'] = df['Clay_Win_Percentage'].str.replace("%","").astype("float")/100

df['Carpet_Matches_Won'] = df['Carpet_Matches_Won'].astype("int64")
df['Carpet_Matches_Lost'] = df['Carpet_Matches_Lost'].str.rstrip(")").astype("int64")
df['Carpet_Win_Percentage'] = df['Carpet_Win_Percentage'].str.replace("%","").astype("float")/100

df['Grass_Matches_Won'] = df['Grass_Matches_Won'].astype("int64")
df['Grass_Matches_Lost'] = df['Grass_Matches_Lost'].str.rstrip(")").astype("int64")
df['Grass_Win_Percentage'] = df['Grass_Win_Percentage'].str.replace("%","").astype("float")/100

df.info()
```

```python
# dropping columns
df = df.drop(['Hard', 'Clay', 'Carpet', 'Grass', 'Hard_Record', 'Clay_Record', 'Carpet_Record', 'Grass_Record'], axis=1)
```
