```python
cormat = df.corr(method='pearson')
round(cormat,2)
```

```python
# Narrow down the attributes (specify which attributes to show in the matrix)
cormat_serve = df[['Current_Rank','1st_Serve_Percentage','1st_Serve_Won_Percentage','1st_Srv._Return_Won_Percentage','2nd_Serve_Won_Percentage','2nd_Srv._Return_Won_Percentage','Ace_Percentage','Ace_Against_Percentage','Break_Points_Ratio','Double_Fault_Percentage','Height_cm']].corr(method='pearson')
sns.heatmap(cormat_serve, annot=True)
plt.show()
```

Intersting insight --> First serve won percentage, and ace percentage has a strong positive correlation with height, but second serve won percentave only has a weak positive correlation with height. 

```python
plot = sns.lmplot(x='1st_Serve_Won_Percentage', y='Total_Win_Percentage', data=df)

plot.set(xlabel = '1st Serve Win %', ylabel = 'Total Win %')
plot.fig.suptitle("Total Win % vs 1st Serve Win %",
                  fontsize=16, fontdict={"weight": "bold"})
```

## https://pbpython.com/pandas-qcut-cut.html --> Creating Bins

```python
df['Quantile_Age'] = pd.qcut(df['Age'], q=4)
```

```python
df.head()
```

```python
df['Quantile_Age'].value_counts()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='Quantile_Age',y='Total_Win_Percentage',data=df, palette='rainbow')
plt.title("Age vs Total Win %")
```

```python
plt.figure(figsize=(8,5))
ax = sns.barplot(x="Quantile_Age", y="Total_Win_Percentage", data=df)
ax.bar_label(ax.containers[0])
plt.title("Age vs Total Win %")
```
